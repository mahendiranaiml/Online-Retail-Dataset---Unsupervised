```python
import pandas as pd

# -----------------------------
# 1. Ensure Date is datetime
# -----------------------------
df['Date'] = pd.to_datetime(df['Date'])

# -----------------------------
# 2. Create Snapshot Date
# -----------------------------
snapshot_date = df['Date'].max() + pd.Timedelta(days=1)

# -----------------------------
# 3. Create RFM Features per Customer
# -----------------------------
rfm = df.groupby('CustomerID').agg({
    'Date': 'max',                 # Last purchase date
    'InvoiceNo': 'nunique',        # Number of orders
    'Spending $': 'sum'            # Total spending
}).reset_index()

# -----------------------------
# 4. Calculate Recency
# -----------------------------
rfm['Recency'] = (snapshot_date - rfm['Date']).dt.days

# -----------------------------
# 5. Rename Columns
# -----------------------------
rfm.rename(columns={
    'InvoiceNo': 'Frequency',
    'Spending $': 'Monetary'
}, inplace=True)

# Keep only useful columns for merge
rfm = rfm[['CustomerID', 'Recency', 'Frequency', 'Monetary']]

# -----------------------------
# 6. Merge Back to Original Data
# -----------------------------
df = df.merge(rfm, on='CustomerID', how='left')

# -----------------------------
# 7. Check Result
# -----------------------------
print(df.head())
```
